```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.array([0, 1, 2, 3, 4, 5, 6, 7]) * 10
Res = [0.45, 0.60, 1.07, 1.87, 2.80, 3.92, 4.37, 4.82]

plt.figure(figsize=(10, 8))
plt.scatter(time, Res, label="applied thermal compound")
plt.axhline(y=12.12, color='r', linestyle='-.', label="without thermal compound")
plt.xlabel("time[min]")
plt.ylabel("Resistance[M Ohm]")
plt.legend()
plt.title("Coupler-Frame resistance after installation")
plt.grid(alpha = 0.3)
```
